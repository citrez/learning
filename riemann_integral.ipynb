## A set of functions for rienmann integrating

```python
a_list = [1,2,3]
list(map(lambda d: d*2 , a_list))
```

```python
from typing import Callable, List
import plotnine as pn
import numpy as np

def plot_function(fn: Callable, domain: List[int] = [-10,10] ):
    x = np.linspace(domain[0],domain[1],10)
    y = list(map(fn, x) )
    return pn.ggplot()+pn.geom_point(pn.aes(x,y))

plot_function(lambda d: d**2, domain=[0,1])
```

```python
def riemann_integral(fn,range,accuracy):
    linsp = np.linspace(range[0],range[1],accuracy,retstep=True)
    x = linsp[0]
    step = linsp[1]
    return sum([j*step for j in map(fn,x)])

riemann_integral(lambda x:x**2, range = [0,1], accuracy=1000)
```
